```python
import spacy
nlp = spacy.load("en_core_web_sm")

import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet')

import textacy
```

## Read book

```python
f= open('/content/harry_potter.txt')
text = f.read()
```

## Segment the text of the book into sentences. How many sentences does this book have?

```python
#Method 1 using Basic method
sentance_list = text.split(".")
print(f"Total Sentences available in harry_potter book is: {len(sentance_list)}")
```

```python
#Method 2 using spacy
doc = nlp(text)
spacy_sentences = list(doc.sents)
print(f"Total Sentences available in harry_potter book is: {len(spacy_sentences)}")
```

```python
#Method 3 using nltk
nltk_sentences = nltk.sent_tokenize(text)
print(f"Total Sentences available in harry_potter book is: {len(nltk_sentences)}")
```

## Compute the frequency of each token in the book. What are the most frequent tokens?

```python
tokens = {}
for i in nltk_sentences:
  sentance_tokens = nltk.tokenize.word_tokenize(i)
  for j in sentance_tokens:
    if j in tokens:
      tokens[j] += 1
    else:
      tokens[j] = 1

print(tokens)

frequent_tokens = sorted(tokens, key=tokens.get, reverse=True)[:20]
for i in frequent_tokens:
  print(i, "\t\t\t",tokens[i])
```

## Choose a sentence from the book. Analyze this chosen sentence.

```python
single_sentence = nltk_sentences[180]
print(single_sentence)
```

### Calculating all n-grams.

```python
#Method 1 (Basic)
def ngrams(input, n):
    input = input.split(' ')
    output = []
    for i in range(len(input)-n+1):
        output.append(input[i:i+n])
    return output

ngrams(single_sentence, 2)
```

```python
#Method 2 using nltk
sentence_tokens = nltk.tokenize.word_tokenize(single_sentence)
ngram = list(nltk.ngrams(sentence_tokens, 3))
print(ngram)
```

```python
#Method 3 using spacy
sentence_doc = nlp(single_sentence)
ngram = list(textacy.extract.basics.ngrams(sentence_doc, 3, filter_stops=False, filter_punct=False))
print(ngram)
```

### Finding POS tags of tokens.

```python
pos_tags = nltk.pos_tag(sentence_tokens)
for t, tag in pos_tags:
  print(t, "\t\t", tag)
```

### Stemming tokens.

```python
porter = nltk.stem.PorterStemmer()
for t in sentence_tokens:
  print(t, "\t\t", porter.stem(t))
```

### Lemmatizing tokens.

```python
lemmatizer = nltk.stem.WordNetLemmatizer()
for t in sentence_tokens:
  print(t, "\t\t", lemmatizer.lemmatize(t))
```
